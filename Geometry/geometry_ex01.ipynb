```python
import folium

m = folium.Map(location=[35.1275, 129.0903], zoom_start=20)
m
```

```python
# tile에 사용가능한 옵션 OpenStreetMap, Stamen Terrain , Stamen Toner
m = folium.Map(location=[35.1175, 129.0903], zoom_start=15,tiles='Stamen Toner' )
m
```

```python
# 부트스트렙을 이용해서 사용가능함
m = folium.Map(location=[35.1175, 129.0903], zoom_start=12)
folium.Marker(location=[35.1175, 129.0903], popup='부경대학교 용당캠퍼스').add_to(m)
folium.Marker(location=[35.178078, 129.199141], popup='송정해수욕장', 
                icon=folium.Icon(color='red', icon='remove-circle')).add_to(m)
m
```

```python
import json

m = folium.Map(location=[35.566345, 126.977893], zoom_start=7) # 한국지도 전체표시
gfile = open('./province-simple.json' , mode='r' , encoding='utf-8').read()
jsondata = json.loads(gfile)
folium.GeoJson(jsondata, name="province").add_to(m)
folium.Marker(location=[35.566345, 126.977893], popup='지도중심').add_to(m)
m
```

```python
m2 = folium.Map(location=[36.505354 , 127.704334], zoom_start=7 , tiles='Cartodb Positron')
point_data = json.loads(open('./point.json', mode='r' , encoding='utf-8').read())

for i in range(len(point_data)): #0~10
    folium.Circle(location=point_data[i], radius=50, color='#000000').add_to(m2)

m2
```
